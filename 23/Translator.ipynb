```python
def prepare_dataset(data):
    res = []
    k = 0
    i = 0
    
    while k < 100:
        if len(data[i]) < 512:
            print("\n short post", k)
            print("Длина: ",len(data[i]))
            print(data[i])
            res.append(data[i])
            k = k + 1
            
        i = i + 1
        
        if i > len(data):
            print("Подключите еще категорию")
            break
        
    return res
```

```python
# 100 предложений для перевода

from sklearn.datasets import fetch_20newsgroups
import numpy as np

categories = ['comp.graphics']

test_b = fetch_20newsgroups(subset='test',
   categories=categories, shuffle=True, random_state=42, remove='headers,footers,quotes')

x_test = test_b.data
data = prepare_dataset(x_test)

print('size of validation set: %s' % (len(data)))
```

```python
from transformers import *

# source & destination languages
src = "en"
dst = "ru"

task_name = f"translation_{src}_to_{dst}"
model_name = f"Helsinki-NLP/opus-mt-{src}-{dst}"

translator  = pipeline(task_name, model=model_name, tokenizer=model_name)
```

```python
for i in range (len(data)):
    print("\n New post: ", i, "\n")
    print(data[i])
    print("\n Перевод:\n")
    print(translator(data[i])[0]["translation_text"])
                
```
